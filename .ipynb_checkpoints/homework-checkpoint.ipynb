```python
import cv2
import time

# 載入影片
cap = cv2.VideoCapture('./video/vtest.avi')

# 取得畫面尺寸
WIDTH = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))       
HEIGHT = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

RECT = ((400, 0), (800, 800))
(left, top), (right, bottom) = RECT


# 取出 ROI 子畫面
def roiarea(frame):                  
    return frame[top:bottom, left:right]
# 將 ROI 區域貼回到原畫面
def replaceroi1(frame, roi):            
    frame[top:bottom, left:right] = roi
    return frame

#存檔並輸出
length = 60.0

fourcc = cv2.VideoWriter_fourcc(*'DIVX')             

size = (int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))

out = cv2.VideoWriter('./video/homework.avi', fourcc, length, size)



frameCount = cap.get(cv2.CAP_PROP_FRAME_COUNT)

fgbg = cv2.createBackgroundSubtractorKNN(detectShadows=True)

index = 1

while True:
    ret, frame = cap.read()
    # if frame is read correctly ret is True
    if not ret:
        print("Can't receive frame . Exiting ...")
        break
    if ret :              
        roi = roiarea(frame)                   # 取出子畫面
#         roi = cv2.cvtColor(roi, cv2.COLOR_BGR2HSV)       # BGR to HSV
        gray = cv2.cvtColor(roi, cv2.COLOR_RGB2GRAY)  #先將RGB轉灰階
        gray = cv2.GaussianBlur(gray, (17, 17), 1)    # 高斯去雜訊
        
        if index>=100  :
            gray = fgbg.apply(gray)
#             cv2.putText(frame, 'KNN', (0, 0), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,0,0), 2, cv2.LINE_AA)
        if index>=300 :
            gray = cv2.Canny(gray, 100, 500)
        roi = cv2.cvtColor(gray, cv2.COLOR_GRAY2BGR)  #將單通道圖像轉換為三通道RGB灰度圖
    #     roi = cv2.Canny(roi, 576, 368)

        frame = replaceroi1(frame, roi)         # 將處理完的子畫面貼回到原本畫面中


        cv2.putText(frame, 'original', (10, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 2, cv2.LINE_AA)
    #     cv2.putText(frame, 'count: ' + str(frameCount), (10, 150), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 2, cv2.LINE_AA)
        cv2.putText(frame, 'count: ' + str(frameCount), (10, 150), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 2, cv2.LINE_AA)
        cv2.putText(frame, 'frame: ' + str(index), (10, 200), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 2, cv2.LINE_AA)
        if index <= 100 :
            cv2.putText(frame, 'ROI+BGRtoGRAY', (500, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,0,0), 2, cv2.LINE_AA)
        elif index <= 300:
            cv2.putText(frame, 'Use KNN', (500, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,0,0), 2, cv2.LINE_AA)
        else :    
            cv2.putText(frame, 'CannyGetOnlyPerson', (450, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,0,0), 2, cv2.LINE_AA)
                 

        index += 1
        
        out.write(frame) 
        cv2.imshow('frame', frame)
        
    
    if cv2.waitKey(90) == 27:
        break

cv2.destroyAllWindows()
cap.release()
```
